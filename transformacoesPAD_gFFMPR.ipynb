## Distribuição das Respostas por Período do dia

Podemos verificar a distribuição de frequências dos respondentes quando ao período do dia que a atividade foi realizada.

```python
import matplotlib.pyplot as plt
```

```python
# Converter data e horarios para datetime
dados1['Data'] = pd.to_datetime(dados1['Data'])

# Separar em períodos do dia (madrugada, manha, tarde, noite)
dados1['Periodos'] = pd.cut(dados1['Data'].dt.hour,
                           bins=[0, 6, 12, 18, 24],
                           labels=['0:00 - 6:00', '6:00 - 12:00',
                                   '12:00 - 18:00', '18:00 - 0:00'])

# Criar figura e eixos
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))
fig.suptitle('Gráfico de Barras das Respostas por Data e Hora - \
             Todas as Turmas')

# Plotar histogramas para cada biblioteca
bibliotecas = ['Numpy', 'Pandas', 'Matplotlib', 'Scikit']

# Iterar sobre os subplots
for i, ax in enumerate(axs.flat):
    biblioteca = bibliotecas[i]
    ax.set_title(biblioteca)
    ax.set_xlabel('Hora')
    ax.set_ylabel('Frequência')

    turmas_data = pd.concat([dados1[dados1['Turma'] == 'T1'],
                             dados1[dados1['Turma'] == 'T2'],
                             dados1[dados1['Turma'] == 'T3']])

    hist_data = turmas_data.groupby('Periodos')[biblioteca].value_counts().unstack().fillna(0)
    hist_data.plot(kind='bar', stacked=False, ax=ax, legend=False, alpha=0.7)


plt.legend(title='Respostas', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

Os dados se referem a todas as turmas e dizem que os respondentes preferem responder ao formulário, majoritariamente, durante o dia, preferencialmente entre o meio-dia e às 18 horas.

```python
nome_arquivo = 'Aquecimento PAD - 2023 (dados agregados) - Dados.csv'
selecao = ['Carimbo de data/hora', 'Vídeos?', '[Numpy]', '[Pandas]',
           '[Matplotlib]', '[Scikit-learn]', 'Equipe']
dados1 = pd.read_csv(nome_arquivo, usecols = selecao)
novos_nomes = {'Carimbo de data/hora': 'Data', 'Vídeos?': 'Videos',
               '[Numpy]': 'Numpy', '[Pandas]': 'Pandas',
               '[Matplotlib]': 'Matplotlib', '[Scikit-learn]': 'Scikit',
               'Equipe': 'Turma'}
dados1 = dados1.rename(novos_nomes, axis = 'columns')
```

```python
from datetime import datetime

# Estabelecendo "Zero" (primeira submissão de cada turma)
t1_0 = datetime(2021,7,31,10,21,32)
t2_0 = datetime(2022,6,3,17,42,30)
t3_0 = datetime(2023,4,28,21,29,24)
# Criando uma coluna Tempo (vazia)
dados1['TTotal'] = ""
dados1['Hora'] = ""
for i in range(len(dados1)):
    tempo = pd.to_datetime( dados1.at[i, 'Data'], format='%d/%m/%Y %H:%M:%S')
    dados1.at[i, 'Hora'] = tempo.hour + tempo.minute / 60
    if dados1.at[i, 'Turma']  == 'T1':
        tempo -= t1_0
    elif dados1.at[i, 'Turma']  == 'T2':
        tempo -= t2_0
    elif dados1.at[i, 'Turma']  == 'T3':
        tempo -= t3_0
    dados1.at[i, 'TTotal'] = tempo.days + tempo.seconds / (24*3600)
```

```python
# Criar um dicionário para mapear as respostas para valores numéricos
resposta_valores = {
    "Nunca ouvi falar!": 0.00,
    "Já ouvi falar!": 0.25,
    """Já "mexi"!""": 0.50,
    "Já trabalhei com ela!": 0.75,
    "Acho que posso ajudar outras pessoas a utilizá-la!": 1.00
}

# Converter a coluna de data e hora para o formato de data/hora
tabela_respostas['Carimbo de data/hora'] = pd.to_datetime(tabela_respostas['Carimbo de data/hora'])

# Criar uma nova coluna com a categoria de horário
tabela_respostas['HoraCategoria'] = pd.cut(tabela_respostas['Carimbo de data/hora'].dt.hour,
                                           bins=[0, 5, 12, 19, 24],
                                           labels=['0:00 - 6:00', '6:00 - 12:00', '12:00 - 18:00', '18:00 - 0:00'])

# Selecionar apenas as colunas relevantes
colunas_relevantes = ['[Numpy]', '[Pandas]', '[Matplotlib]', '[Scikit-learn]']


# Criar uma nova coluna com os valores numéricos das respostas para cada coluna
for coluna in colunas_relevantes:
    tabela_respostas[coluna + '_Valores'] = tabela_respostas[coluna].map(resposta_valores)

# Calcular a média dos valores numéricos para cada coluna
medias_por_coluna = {}
for coluna in colunas_relevantes:
    medias_por_coluna[coluna] = tabela_respostas[coluna + '_Valores'].mean()

# Cores
cores = ['#FFA500', '#005AFF', '#00FF57', '#A500FF']

# Plotar os gráficos separados por horários
for i, coluna in enumerate(colunas_relevantes):
    plt.figure()
    tabela_respostas.groupby('HoraCategoria')[coluna + '_Valores'].mean().plot(kind='bar', color=cores[i])
    plt.title(coluna)
    plt.xlabel('Faixa de Horário')
    plt.ylabel('Média')
    plt.show()
```
